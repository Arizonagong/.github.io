```python
import pandas as pd
import pandas_alive
```

```python
# Read the csv file
df=pd.read_csv("covid19.csv")

# Change "data_reported" column into "datetime" data format
df['Date_reported'] = pd.to_datetime(df.Date_reported)

# Data pre-precessing

# Check up the duplicates at the entire table level
duplicateRows = df[df.duplicated()]
duplicateRows

# And drop the duplicated rows
df=df.drop_duplicates()

# Check up the duplicates at specific column combination.
df=df.drop_duplicates(subset=['Date_reported','Country'])

# And drop the duplicated rows one more time
df.dropna(axis=0)
df

# Check whether there is remaining missing values
df['Date_reported'].isnull().values.any()

# Converting long format data into short format
df = pd.pivot(df, 
              index='Date_reported', # Targeted index column: It should be the date column.
              columns='Country', # The column that will be transposed into column names with its unique values.
              values='Cumulative_cases') # The column that will be the value of each columns and rows.

# Check up how your data was converted.
df = df.iloc[:, 10:25] # I chose only specific number of rows as the size if data is limited in this computer.
df.tail()
```
