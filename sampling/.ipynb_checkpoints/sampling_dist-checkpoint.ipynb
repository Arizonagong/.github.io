```python
import pandas as pd
import numpy as np
import matplotlib as map
import matplotlib.pyplot as plt
import random
```

```python
df = pd.read_csv('EAI.csv')
print(df)
type(df)
df.describe()
df.columns = ["manager", "annual_salary","training_program"]

print(df)
```

```python

n=30
num=3
lst1=[]
lst2=[]
for i in range(num):
    df_sampl=df.sample(n=n)
    sample=list(df_sampl["annual_salary"])
    lst1.append(sample)
    sample_m=df_sampl["annual_salary"].mean()
    lst2.append(sample_m)
print(lst1)
```

```python
x1=lst1[0]
x2=lst2
n_bins = 15
print(x2)
```

```python
plt.hist(x1, n_bins, histtype='bar', stacked=True, rwidth=0.9,
         color="red")
plt.ylim(0, n*0.5)
plt.axvline(x = x2[len(x2)-1], color = 'b')
plt.text(x2[len(x2)-1],n*0.5/2,f'{x2[len(x2)-1]}',rotation=90)
plt.show()
```

```python
plt.hist(x2, 
         n_bins, 
         histtype='bar', 
         stacked=True, 
         rwidth=0.9,
         color="blue")
plt.ylim(0, num*0.5)
plt.show()
```
