```python
import pandas as pd
import numpy as np
import matplotlib as map
import matplotlib.pyplot as plt
import streamlit as slt
```

```python
df = pd.read_csv('EAI.csv')
print(df)
type(df)

df.columns = ["manager", "annual_salary","training_program"]
df.describe()
```

```python

n=30
num=100
lst1=[]
lst2=[]
for i in range(num):
    df_sampl=df.sample(n=n)
    sample=list(df_sampl["annual_salary"])
    lst1.append(sample)
    sample_m=df_sampl["annual_salary"].mean()
    lst2.append(sample_m)
print(lst2)
```

```python
x1=lst1[1]
x2=lst2
n_bins = 15
print(x2)
```
