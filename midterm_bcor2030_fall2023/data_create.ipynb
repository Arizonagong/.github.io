```python
import numpy as np
import pandas as pd
```

```python
# Daily water intake amount data for the Songdo adults

    # Each observation: one person
    # Mean: liter per day
    # SD: liter per day
    # size: 90,000 Songdo residents installed
dat_wt=np.random.normal(loc=2.5, scale=0.3, size=90000)
```

```python
df_wt=pd.DataFrame(dat_wt, columns=['wt_day'])
df_wt
df_wt.to_csv("water.csv")
```

```python
# Life expectancy data

# 1. College Graduate
    # Each observation: one person
    # Mean: age of death
    # SD: 
    # size: 100,000
dat_lfexp1=np.random.normal(loc=85, scale=3.6, size=100000)
dat_lfexp1=pd.DataFrame(dat_lfexp1, columns=['age_of_death'])
dat_lfexp1.describe()
```

```python
# Average heat of Aphone 15

    # Each observation: one Cell phone
    # Mean: heat by celsius
    # SD: heat by celsius
    # size: 500
dat_aphone=np.random.normal(loc=45, scale=2, size=500)
dat_aphone
```

```python
df_aphone=pd.DataFrame(dat_aphone, columns=['heat'])
df_aphone=df_aphone.reset_index(drop=True)
df_aphone.to_csv("aphone.csv")
```

```python
# Average GPA for DESB

    # Each observation: one student
    # Mean: GPA(0-4)
    # SD: GPA
    # size: 2500
dat_gpa=np.random.normal(loc=3.2, scale=0.5, size=2500)
dat_gpa
```

```python
df_gpa=pd.DataFrame(dat_gpa, columns=['gpa'])
df_gpa=df_gpa.reset_index(drop=True)
df_gpa.to_csv("gpa.csv")
```
