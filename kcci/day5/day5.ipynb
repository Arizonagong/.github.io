# Titanic Analysis

```python
import pandas as pd
import seaborn as sns
# %%
titanic=sns.load_dataset('titanic')
titanic.head()
# %%
# Countplot
sns.catplot(kind='count',x ="alone", hue ="survived", dodge=True, data = titanic)
```

```python
sns.catplot(
    data=titanic, x="pclass", y="survived", hue="alone",
    kind="point"
)
```

```python
sns.catplot(data=titanic, x='alone', hue='survived', kind='count')
```

```python
sns.catplot(
    data=titanic,
    x="sex", col="class",
    kind="count", orient="v",
    sharex=False, margin_titles=True,
    height=9, aspect=0.5,
)
```
